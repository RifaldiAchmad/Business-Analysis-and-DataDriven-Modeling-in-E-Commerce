### Explore products and product_category data

```python
product_all = products.merge(product_category, on='product_category_name', how='inner')


columns_to_drop = ['product_category_name']

product_all = product_all.drop(columns=columns_to_drop)
product_all.head()
```

```python
# Add weight category column
def weight_category(weight):
    if weight <= 500:
        return 'Light'
    elif weight <= 2000:
        return 'Medium'
    else:
        return 'Heavy'
product_all['weight_category'] = product_all['product_weight_g'].apply(weight_category)

# Added dimension category columns based on length, width, and height
def size_category(length, width, height):
    volume = length * width * height
    if volume <= 1000:
        return 'Small'
    elif volume <= 5000:
        return 'Medium'
    else:
        return 'Large'
product_all['size_category'] = product_all.apply(
    lambda row: size_category(row['product_length_cm'], row['product_width_cm'], row['product_height_cm']), axis=1
)

product_all['weight-size'] = product_all['weight_category'] + '-' + product_all['size_category']
product_all.head()
```

### Pertanyaan 2: What is the sales and revenue trend for the best-selling products on the best sellers in the last year?

```python
# Filter data only from the last year
last_year = best_seller['order_purchase_timestamp'].max() - pd.DateOffset(years=1)
filtered_data = best_seller[best_seller['order_purchase_timestamp'] > last_year]

# Take the most categories
top_category = filtered_data['product_category_name_english'].value_counts().idxmax()
filtered_top_category = filtered_data[filtered_data['product_category_name_english'] == top_category]

# Group monthly data for orders and sales
monthly_data = filtered_top_category.groupby(filtered_top_category['order_purchase_timestamp'].dt.to_period('M')).agg(
    total_order=('order_id', 'count'),
    total_sales=('price', 'sum')
)

# Plot
fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

for i, (col, ylabel) in enumerate(zip(['total_order', 'total_sales'], ['Total Orders', 'Total Sales ($)'])):
    sns.lineplot(data=monthly_data, x=monthly_data.index.to_timestamp(), y=col, marker='o', ax=axes[i])
    axes[i].set_ylabel(ylabel)
    axes[i].set_title('Total Orders and Total Sales of Best-Selling Products in the Last Year' if i == 0 else '', fontsize=16)
    axes[i].grid(True, linewidth=0.3, alpha=0.5)
    axes[i].spines[['top', 'right']].set_visible(False)
    axes[i].set_xlabel('')  # Menghapus label di sumbu x
    for x, y in zip(monthly_data.index.to_timestamp(), monthly_data[col]):
        axes[i].text(x, y, f'{y:,.0f}', fontsize=8, ha='center', va='bottom', color='black')

    axes[i].minorticks_on()
    axes[i].grid(linewidth=0.2, linestyle='--', which='minor', alpha=0.3)

plt.tight_layout()
plt.subplots_adjust(hspace=0.1)
plt.show()
```

**Insight:**
- The growth in the number of orders and total sales over the past year for the best sellers tends to increase
- The smallest number was around mid-2017 and the peak was in mid-2018

## Analisis Lanjutan (Customer Segmentation Based on RFM score)

```python
rfm_df = best_seller.groupby("customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max", # take last order date
    "order_id": "nunique", # calculate the number of orders
    "price": "sum" # calculate the amount of revenue generated
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

# Calculate when the customer last made a transaction (day)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date

recent_date = best_seller["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.sort_values("monetary", ascending=False).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 12))

# Recency plot
max_recency = rfm_df["recency"].max() * 1.1  # Tambahkan 10% margin
sns.barplot(
    y="recency",
    x="customer_id",
    data=rfm_df.sort_values("recency", ascending=False).head(5),
    ax=ax[0]
)
ax[0].set_ylim(0, max_recency)
ax[0].set_ylabel(None)
ax[0].set_xlabel("Customer ID")
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis="x", labelsize=15, rotation=90)

# Frequency plot
max_frequency = rfm_df["frequency"].max() * 1.1  # Tambahkan 10% margin
sns.barplot(
    y="frequency",
    x="customer_id",
    data=rfm_df.sort_values("frequency", ascending=False).head(5),
    ax=ax[1]
)
ax[1].set_ylim(0, max_frequency)
ax[1].set_ylabel(None)
ax[1].set_xlabel("Customer ID")
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis="x", labelsize=15, rotation=90)

# Monetary plot
max_monetary = rfm_df["monetary"].max() * 1.1  # Tambahkan 10% margin
sns.barplot(
    y="monetary",
    x="customer_id",
    data=rfm_df.sort_values("monetary", ascending=False).head(5),
    ax=ax[2]
)
ax[2].set_ylim(0, max_monetary)
ax[2].set_ylabel(None)
ax[2].set_xlabel("Customer ID")
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis="x", labelsize=15, rotation=90)

# Set overall title
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=25, y=1.03)

plt.tight_layout()
plt.show()
```

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df
```

```python
# Normalizing the rank of the customers
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100

rfm_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)
rfm_df
```

```python
rfm_df['RFM_score'] = 0.15 * rfm_df['r_rank_norm'] + 0.28 * rfm_df['f_rank_norm'] + 0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_id', 'RFM_score']].sort_values('RFM_score', ascending=False).reset_index(drop=True)
```

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

rfm_df[['customer_id', 'RFM_score', 'customer_segment']]
```

```python
customer_segment_df = rfm_df.groupby("customer_segment", as_index=False)["customer_id"].nunique().sort_values("customer_id", ascending=False).reset_index(drop=True)
customer_segment_df
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])

plt.figure(figsize=(10, 5))
colors_5 = ["#D3D3D3", "#72BCD4", "#72BCD4", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    x="customer_id",
    y="customer_segment",
    data=customer_segment_df.sort_values(by="customer_segment", ascending=False),
    palette=colors_5
)
plt.title("Number of Customer for Each Segment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

**Insight:**
- Customers with the longest last order time, which is around 550 days
- The frequency of each customer's order is only 1 time
- The biggest revenue was earned by one customer, which was almost $1200
- The customers with the largest category are low value customers, followed by medium value customers
- Top customers no more than 100 people
